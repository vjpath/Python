##### Cleaning the data

```python
# Rename the column

cars = cars.rename(columns={'Unnamed: 1':'model'})
#cars
```

##### Droping unwanted columns from dataset

```python
cars=cars.drop(columns=['drat'])
cars.head()
```

##### Replacing the missing values with mean of that column

```python
# Missing values in qsec column is replaced by qsec mean values
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
cars.head()
```

```python
cars.mpg = cars.mpg.astype(int)
cars.dtypes
```

```python
# Sorting values of 'cyl' column in ascending order

cars.sort_values(by='cyl')
```

```python
#Filter records based on condition where cyl should be greater than 6
Filt1 = cars['cyl'] > 6 
Filt_new = cars[Filt1]
Filt_new
```

```python
#Filter using multiple conditions
Filt2 = (cars['cyl'] > 6) & (cars['hp'] > 300)
Filt_new = cars[Filt2]
Filt_new
```

```python
# group by
cars.groupby('cyl')['mpg'].mean()
```
